# El Niño and the Southern Oscillation (Satatistical analysis)

The goal of this project is to use machine-learning techniques (train an xgboost regression model) to predict the values of the measurements of the Southern Oscillation Index (SOI) for the following few months. These indexes are strongly associated with El niño phenomenon.

GGiven the nature of the datum, I decided to make time series forecasting using machine learning tools, namely the xgboost regression python tool. So, we transform the time series forecasting into a regression problem.

```python
#Import the usual libraries 
import pandas as pd   #pandas for data 
import numpy as np #pandas for numerical analysis
import matplotlib.pyplot as plt #plot data 
import seaborn as sns #color pallete 
import xgboost as xgb #Time Series Forecasting ML model
from sklearn.metrics import mean_squared_error #error metrics 
color_pal = sns.color_palette() #setting color pallete
plt.style.use('fivethirtyeight') #setting plot style
```

# 1. Data Pre-Processing

We aplay certain data frame manipulations in order to give our main data sets a time seres like strustrure.

```python
#data frame of the standarized vaules of the  Southern Oscillation Index (SOI) 
soi_s = pd.read_csv("soi_stand.csv",index_col=0)
#data frame of the standarized vaules Outgoing Longwave Radiation (OLR)  
orl_s = pd.read_csv("orl_stand.csv",index_col=0)
```

```python
# This code transforms  each datasets soi_s and orl_s into one list of its elements 
# (by row and following the order the index order)
# output: soi_col list thti the elemts of the dataset soi_s and orl_col list thti the elemts of the dataset orl_s 

soi_col = [] # empty list to be filled with the soi_s elements 
orl_col = [np.nan] * 276 # list filled with 276 nan in order to compensate the missing dates in orl_s

# loop that runs over the rows of soi_s and append each element to soi_col

for row in soi_s.values.tolist():
    for item in row:
        soi_col.append(item)
        
# loop that runs over the rows of orls_s and append each element to orl_col
for row in orl_s.values.tolist():
    for item in row:
        orl_col.append(item)
```

```python
#output: series time dataset with the SOI and OLR measurements indexed with the date. 

#date range since JAN of 1951 to SEP of 2023
date_range = pd.date_range(start="1951-01-01", end="2023-12-31", freq='M')

# extraxt the year and month of each date
year_month_index = [date.strftime('%Y-%m') for date in date_range]

# The list off measuremnts for each date
data = {
    'SOI': soi_col,
    'ORL': orl_col
}

# give the dataset with the SOI and OLR measurements indexed with the date
ts = pd.DataFrame(data, index=year_month_index)
```

```python
ts.index = pd.to_datetime(ts.index) #in order to extrac the data we need to make sure that the date index have the correct format
ts.index #plot the indexes and we confirme that the data type is a date 
```

## Data visualization
From a visual inspection we can see that the SOI and ORL measurments haver a hight digree of similarity. 

```python
ts.plot(style='-',
        lw=1.5,
        figsize=(30, 5),
        color=color_pal,
        title='Values of SOI and ORL')
plt.show()
```

```python
ts.plot(kind='hist', bins=500) #in orden to find outlier data we plan a frequency histogram
```

```python
ts.query('SOI < -3')['SOI'] \
    .plot(style='.',
          figsize=(15, 5),
          color=color_pal[5],
          title='Outliers')
#these is the evidence of potential outlier data
```

```python
ts = ts.query('SOI > -3').copy() #eliminate outlier data 
```

## 4. Hyper Parameter Optimization

```python
# By using the followin function we extract the information contained in the date index:
# the function add the features cointaned in the date, namely the year and month 
def create_features(df):
    """
    Create time series features based on time series index.
    """
    df = df.copy()
    df['month'] = df.index.month
    df['year'] = df.index.year
    return df
ts = create_features(ts)
```

## 2. Train-Test Split 
Instead of taking random samples, we establish a year threshold to select the train data. Even though these elections might seem arbitrary we can give the following justifications.
1. Remember that the goal is to predict information for the coming months, so it makes sense to train the model with information from the past instead of a random sample.
2. Experimenting with the values of the model, I observed that the smallest errors are obtained with a treshold between 2013 and 2017.

```python
year_threshold=2017 # fixed threshold
```

```python
ts_train=ts[ts['year']<year_threshold]
X_train=ts_train.drop(['SOI'],axis=1)  
y_train=ts_train['SOI'] #target 
```

```python
ts_test=ts[ts['year'] >= year_threshold]
X_test=ts_test.drop(['SOI'],axis=1)
y_test=ts_test['SOI'] #target 
```

## 3. Calling the XGBoost Regressor
We call the XGBRegressor tool from xgboost an obtain rough prediction. We latter adjust the hyperpamerts by the GridSearchCV method from sklearn.

```python
reg = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=1000) 
# tran the model
reg.fit(X_train, y_train,
        verbose=True)  # Change verbose to False if you do not want to see it train
yhat = reg.predict(X_test) # test the model
```

```python
from sklearn.metrics import mean_squared_error as MSE #as we eliminate the apparent oustanding data makes sense to use the mean squeare error metric
rmse = np.sqrt(MSE(y_test, yhat))
print("RMSE : % f" %(rmse))
```

```python
plt.plot(y_test.values, label='Original')
plt.plot(yhat, color='red', label='XGboost')
plt.legend()
```

We can se that the model fits for short times

## 4.Hyper Parameter Optimization
We call the GridSearchCV method from sklearn wich exhaustive search over specified parameter values for an estimator. See [the documentation](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html)

The goal is to obtain a  improvement in the error.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score
params = {
    'min_child_weight': [4, 5],
    'gamma': [i / 10.0 for i in range(3, 6)],
    'subsample': [i / 10.0 for i in range(6, 11)],
    'colsample_bytree': [i / 10.0 for i in range(6, 11)],
    'max_depth': [2, 3, 4,5 ,6]
}
# Initialize XGB and GridSearch
xgb_reg = xgb.XGBRegressor(nthread=-1, objective='reg:squarederror')
grid = GridSearchCV(xgb_reg, params)
grid.fit(X_train, y_train)
gridcv_xgb = grid.best_estimator_
print(r2_score(y_test, gridcv_xgb.predict(X_test)))
```

We obtained a significant improvement in the error!

```python
from xgboost import plot_importance
_ = plot_importance(gridcv_xgb, height=1.5)
```

```python
gridcv_xgb #we see the especifications of the model, personaly I expected a gradient boost but that is not the case fot the data.
```

```python
yhat = grid.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error as MSE
rmse = np.sqrt(MSE(y_test, yhat))
print("RMSE : % f" %(rmse))
```

```python
plt.plot(y_test.values, label='Original')
plt.plot(yhat, color='red', label='XGboost')
plt.legend()
```

We see that the model fits a little better!

## 5. Results and conclusion
 In this final section, we obtain the predictions for the following months: from September 2023 to April 2024.

```python
# Create future dataframe
future = pd.date_range('2023-09-01','2024-06-01', freq='1M') # date index for the future months 
future_ts = pd.DataFrame(index=future)
future_ts['isFuture'] = True
ts['isFuture'] = False 
ts_and_future = pd.concat([ts, future_ts])  # we combine the given data frame with the informations form the future
ts_and_future = create_features(ts_and_future)
```

```python
future_w_features = ts_and_future.query('isFuture').copy()
```

```python
X_future=future_w_features.drop(['SOI','isFuture'],axis=1) #Select the features to feed our model and obtain the predictions
```

```python
soi_pred = grid.predict(X_future) #   obtain the predictions
ts_prediction=pd.DataFrame(index=future)
ts_prediction['SOI_pred']=soi_pred
ts_prediction #  print the predictions
```

```python
#  plot the predictions
ts_prediction.plot(figsize=(10, 5),
                               color=color_pal[4],
                               ms=1,
                               lw=1,
                               title='Future Predictions')
plt.show()
```

### Conclusion 

We predict prolonged periods of negative SOI values. We can expect El Niño to occur during the first month of 2024, maybe reaching its peak by April. These predictions agree with the [ENSO Alert System Status](https://www.cpc.ncep.noaa.gov/products/analysis_monitoring/enso_advisory/ensodisc.shtml).

## 6. Optional: Another approach to Hyper Parameter Optimization.

This is the firts tiem I use these technique but I want to share with you the results.

Instead of using the GridSearchCV method from sklearn we can train the set with diferent sub data set in order to find better parameters, as I show below.

IMPORTANT: I can't tell you which approach is the best, I still explain the code to you. Personally, I would opt for the forts one because it is much clearer how it works for me.

```python
#restar the original form of the ts data set 

data = {
    'SOI': soi_col,
    'ORL': orl_col
}

# give the dataset with the SOI and OLR measurements indexed with the date
ts = pd.DataFrame(data, index=year_month_index)
ts.index = pd.to_datetime(ts.index)
ts = ts.query('SOI > -3').copy() #eliminate outlier data 
ts #see the original data frame
```

```python
#Split the series into 5 trainig subsets

from sklearn.model_selection import TimeSeriesSplit 

tss = TimeSeriesSplit(n_splits=5, test_size=12*10, gap=1)
ts = ts.sort_index()
```

```python
#Plot the  5 trainig subsets
fig, axs = plt.subplots(5, 1, figsize=(20, 15), sharex=True)

fold = 0
for train_idx, val_idx in tss.split(ts):
    train = ts.iloc[train_idx]
    test = ts.iloc[val_idx]
    train.plot(ax=axs[fold],
                          label='Training Set',
                          title=f'Data Train/Test Split Fold {fold}')
    test.plot(ax=axs[fold],
                         label='Test Set')
    axs[fold].axvline(test.index.min(), color='black', ls='--')
    fold += 1
plt.show()
```

```python
# Again we append the feaures of the date 
def create_features(df):
    """
    Create time series features based on time series index.
    """
    df = df.copy()
    df['month'] = df.index.month
    df['year'] = df.index.year
    return df
ts = create_features(ts)
```

```python
#Train the model with 5 iteraations using the 5 data training sets.
tss = TimeSeriesSplit(n_splits=5, test_size=12*10, gap=1)
ts = ts.sort_index()

fold = 0
preds = []
scores = []
for train_idx, val_idx in tss.split(ts):
    train = ts.iloc[train_idx]
    test = ts.iloc[val_idx]

    train = create_features(train)
    test = create_features(test)

    FEATURES = [ 'ORL', 'month', 'year']
    TARGET = 'SOI'

    X_train = train[FEATURES]
    y_train = train[TARGET]

    X_test = test[FEATURES]
    y_test = test[TARGET]

    reg = xgb.XGBRegressor(base_score=0.5, booster='gbtree',    
                           n_estimators=1000,
                           early_stopping_rounds=50,
                           objective='reg:squarederror',
                           max_depth=3,
                           learning_rate=0.01)
    reg.fit(X_train, y_train,
            eval_set=[(X_train, y_train), (X_test, y_test)],
            verbose=100)

    y_pred = reg.predict(X_test)
    preds.append(y_pred)
    score = np.sqrt(mean_squared_error(y_test, y_pred))
    scores.append(score)
```

```python
print(f'Score across folds {np.mean(scores):0.4f}') #obtaining
print(f'Fold scores:{scores}')
```

```python
# Retrain on all data
ts = create_features(ts)

FEATURES = ['ORL', 'month', 'year']
TARGET = 'SOI'

X_all = ts[FEATURES]
y_all = ts[TARGET]

reg = xgb.XGBRegressor(base_score=0.5,
                       booster='gbtree',    
                       n_estimators=1000,
                       objective='reg:squarederror',
                       max_depth=3,
                       learning_rate=0.01)
reg.fit(X_all, y_all,
        eval_set=[(X_all, y_all)],
        verbose=100)
```

```python
# Create future dataframe in order to feed the model to obtein predictions.
future = pd.date_range('2023-09-01','2024-12-31', freq='M')
future_ts = pd.DataFrame(index=future)
future_ts['isFuture'] = True
ts['isFuture'] = False
ts_and_future = pd.concat([ts, future_ts])
ts_and_future = create_features(ts_and_future)
future_w_features = ts_and_future.query('isFuture').copy()
```

```python
future_w_features['pred'] = reg.predict(future_w_features[FEATURES])  #obtain the prediction with the model 
```
